# Distplot numerical -categorical

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False,label='Not survived')

sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False,label='Survived')
```

```python
import matplotlib.pyplot as plt
sns.kdeplot(titanic[titanic['Survived']==0]['Age'], label='Not Survived')
sns.kdeplot(titanic[titanic['Survived']==1]['Age'], label='Survived')
plt.legend()
```

## heatmap categorical -categorical

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
survival_pct = titanic.groupby('Pclass')['Survived'].mean().mul(100).round(2).plot(kind='bar')
print(survival_pct)
```

```python
titanic.groupby('Sex')['Survived'].mean().plot(kind='bar')
```
