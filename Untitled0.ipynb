```python
import pandas as pd
from sklearn import datasets
iris=datasets.load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
df.head()
```

```python
df['target']=iris.target
df.head()
```

```python
iris.target_names
```

```python
df[df.target==2].head()
```

```python
df['flower_name']=df.target.apply(lambda x:iris.target_names[x])
df.head()
```
